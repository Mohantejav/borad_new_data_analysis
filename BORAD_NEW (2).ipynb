```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(data['Country'], data['Value'], color='skyblue')
plt.title('Mobile Data Usage per Mobile Broadband Subscription in 2012')
plt.xlabel('Country')
plt.ylabel('Mobile Data Usage (Units)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#scatter
plt.figure(figsize=(10, 6))
plt.scatter(data['Country'], data['Value'], color='skyblue', alpha=0.7)
plt.title('Mobile Data Usage by Country in 2012')
plt.xlabel('Country')
plt.ylabel('Mobile Data Usage (Units)')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Value'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Mobile Data Usage in 2012')
plt.xlabel('Mobile Data Usage (Units)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```
